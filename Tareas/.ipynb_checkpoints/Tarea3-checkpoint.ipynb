```python
%%time

## Common way to solve the problem:
# import numpy as np
# n = 100
# a = np.zeros((n,))
# a[0:2] = [0,1]
# for i in range(2,n):
#     a[i] = (i)*a[i-2] + (i)*a[i-1] 



## Uncommon way to solve the problem: 
import numpy as np

def with_functions(n, a0, a1):

    a = np.zeros((n,))
    a[0:2] = [a0, a1]
        
    def llenar_vector(i):
        nonlocal a
        a[i] = (i)*a[i-2] + (i)*a[i-1] 
        
    [llenar_vector(i) for i in range(2, n)]
    

    return a

n = 100
with_functions(n, 0, 1)
```
